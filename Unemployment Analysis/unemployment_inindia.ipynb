# PIE PLOTS

PARTITION ACCORDING TO AREA on "Unemployment in india" dataset

```python
fig, ax = plt.subplots(figsize =(8, 7))
#labels
m=['Rural','Urban']
#title for chart
ax.set_title("\nPARTITION ACCORDING TO AREA IN \"Unemployment in india\" dataset",color="brown")
# Creating color parameters
colors = ( "blue" ,"magenta")
# Creating explode data
explode = (0.2, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating plot
wedges, texts, autotexts = ax.pie(data_set2['Area'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 105,
                                  wedgeprops = wp,
                                  textprops = dict(color ="brown"))
# Adding legend
ax.legend(wedges,m,
          title ="AREA",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```

PARTITION ACCORDING TO REGION DIRECTIONS IN "Unemployment_Rate_upto_11_2020" dataset

```python
data_set1['Region.1'].unique()
```

```python
fig, ax = plt.subplots(figsize =(8, 7))
#labels
m=['South', 'Northeast', 'East', 'West', 'North']
#title for chart
ax.set_title("\nPARTITION ACCORDING TO REGION DIRECTIONS",color="brown")
# Creating color parameters
colors = ( "orange", "cyan", "brown","grey", "indigo")
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating plot
wedges, texts, autotexts = ax.pie(data_set1['Region.1'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
# Adding legend
ax.legend(wedges,m,
          title ="Directions",
          loc ="center left",
          bbox_to_anchor =(1.2, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```

# Bar Plot - Average Unemployment Rate by Region:


```python
avg_unemployment_by_region = data_set1.groupby('Region.1')[' Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(10, 6))
avg_unemployment_by_region.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# Scatter Plot - Estimated Unemployment Rate vs. Estimated Employed:

```python
plt.figure(figsize=(8, 6))

# Example scatter plot between unemployment rate and employed individuals
plt.scatter(data_set1[' Estimated Employed'], data_set1[' Estimated Unemployment Rate (%)'], alpha=0.5)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Scatter Plot: Unemployment Rate vs. Employed')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Box Plot - Unemployment Rate by Frequency:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=' Frequency', y=' Estimated Unemployment Rate (%)', data=data_set1)
plt.xlabel('Frequency')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Box Plot: Unemployment Rate by Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
sns.pairplot(data_set1[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']])
plt.tight_layout()
plt.show()
```

# Violin Plot - Unemployment Rate by Region:

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Region.1', y=' Estimated Unemployment Rate (%)', data=data_set1)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Violin Plot: Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Histogram - Distribution of Unemployment Rate:

```python
plt.figure(figsize=(10, 6))
plt.hist(data_set1[' Estimated Unemployment Rate (%)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title('Histogram: Distribution of Unemployment Rate')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_set1[' Estimated Employed'], data_set1[' Estimated Unemployment Rate (%)'], data_set1[' Estimated Labour Participation Rate (%)'], c='blue', marker='o')

# Set labels for each axis
ax.set_xlabel('Estimated Employed')
ax.set_ylabel('Estimated Unemployment Rate (%)')
ax.set_zlabel('Estimated Labour Participation Rate (%)')

# Set the title
ax.set_title('3D Scatter Plot')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D line plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for region in data_set1['Region.1'].unique():
    region_data = data_set1[data_set1['Region.1'] == region]
    ax.plot(region_data[' Estimated Employed'], region_data[' Estimated Unemployment Rate (%)'], region_data[' Estimated Labour Participation Rate (%)'], label=region)

# Set labels for each axis
ax.set_xlabel('Estimated Employed')
ax.set_ylabel('Estimated Unemployment Rate (%)')
ax.set_zlabel('Estimated Labour Participation Rate (%)')

# Set the title
ax.set_title('3D Line Plot')

plt.legend()
plt.tight_layout()
plt.show()
```
